```python
# Import libraries:
import numpy as np
import pandas as pd
import seaborn as sns  # optional, but recommended
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
# Add 'diagnosis' key for convenience
df['diagnosis'] = np.array([cancer.target_names[i] for i in cancer.target])

# Plot a scatter plot of 'mean compactness' and 'mean perimeter'
sns.scatterplot(x='mean compactness', y='mean perimeter', hue='diagnosis', data=df)
```

```python
# A pairwise scatter plot of 'mean texture', 'mean perimeter', 'mean compactness', and 'mean radius'
sns.pairplot(df, vars=['mean texture', 'mean perimeter', 'mean compactness', 'mean radius'], hue='diagnosis')
```

#### Do you notice that two of these features seem to be particularly correlated? Why might this be?

"mean perimeter" and "mean radius" are highly correlated. This is because of the mathematical relationship of radius and perimeter. Denoting the perimiter as $p$ and the radius as $r$, $p = 2r\pi$. Likewise, $r=\frac{p}{2\pi}$
